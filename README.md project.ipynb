#Traffic Accident Severity Prediction
PROJECT OVERVIEW:
This project is aimed to predict wether a traffic accident is severe or not severe in a range of 0 and 1

It follows a well structured 5- week DSA project guideline
- Week 1: Introduction & Dataset Loading  
- Week 2: Data Cleaning & Exploration  
- Week 3: Feature Engineering & Selection  
- Week 4: Model Building  
- Week 5: Model Evaluation & Reporting  

DATASET DESCRIPTION:
The dataset used in this prediction named dataset_traffic_accident_prediction which is found on kaggle.
The dataset contains various factors sorrounding causes of accidents including the follwowing.
1. Environmental factors which include; weather condition, light condition and visibility.
2. Road features which include; road type, road surface.
3. Driver and vehicle features which include; driver's age, vehicle type and number of vehicles involved etc.
4. Time which includes; day of the week or time of the day.
The target variable in this dataset is accident severity which is classified into two categories: severe and not severe.
where: 0 = not severe and 1= severe.

TECHNOLOGIES USED ARE:
-Python, pandas,numpy,matplotlib, seaborn, scikit-learn,jupyter notebook and statsmodel.

FEAUTURES USED AND PERFORMED IN THIS PROJECT
- Data Cleaning: Handling missing values, removing duplicates, correcting data types.
- Exploratory Data Analysis (EDA): Visualizing data distributions, correlations, and patterns.
- Feature Engineering and Selection: Creating new features, encoding categorical variables, scaling numerical features.
- Model Building: Implementing various machine learning algorithms (Logistic Regression, Decision Trees, Random Forest, etc.)
- Model Evaluation: Using metrics like accuracy, precision, recall, F1-score.
The performance was as follows:
| Model              | Accuracy | Precision | Recall | F1-Score |
|--------------------|----------|-----------|--------|----------|
| Logistic Regression| 0.68     | 0.45      | 0.10   | 0.16     |



PROJECT OUTCOME:
The project successfully developed a predictive model that can classify traffic accidents into severe and not severe categories with a high degree of accuracy. The insights gained from the analysis can help in understanding the factors contributing to accident severity, which can be valuable for policymakers and traffic safety authorities in implementing preventive measures.
