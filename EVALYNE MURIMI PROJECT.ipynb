```python
#WEEK 3:FEATURE ENGINNERING AND SELECTION
df = pd.read_csv(r"C:\Users\Admin\Downloads\dataset_traffic_accident_prediction1.csv")
#Objective: Create new features and remove irrelevant ones.
#creating meaningful features
# create binary indicator for alcohol involvement using existing 'Driver_Alcohol' column
df['Alcohol involvement'] = (df['Driver_Alcohol'] > 0).astype(int)
print(df[['Driver_Alcohol', 'Alcohol involvement']].head())
```

```python
#removing irrelevant features
cols_to_drop = [col for col in ['Driver_Alcohol', 'Accident_ID', 'Date', 'Time'] if col in df.columns]
df = df.drop(columns=cols_to_drop)
print(df.head())
```
